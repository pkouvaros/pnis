```python
from GuardSwarmDilemma import Simulation
import pandas as pd
import numpy as np
```

```python
sim = Simulation(number_of_agents=10)
# sim = Simulation()
```

```python
from matplotlib import pyplot as plt


for i in range(20):
    policy_outputs = pd.DataFrame(sim.template_agent.strategy.policy_net.predict(  # type: ignore
        np.arange(0, 1, 0.01)), index=np.arange(0, 1, 0.01))  # type: ignore
    policy_outputs.plot(xlabel="Normalised HP", ylabel="Q-Value")
    plt.show(block=False)
    plt.savefig(f"Q-Functions/anim/{i}.png")

    sim.train(1)
```
